### missing 데이터 확인

```python
import pandas as pd
import numpy as np
# 초기 df 설정
dates = pd.date_range('20130101', periods=6)
df = pd.DataFrame(np.random.randn(6, 5), index=dates, columns=list('ABCDF'))
df.head()
```

```python
df1 = df.reindex(index = dates[0:4], columns = list(df.columns) + ['E'])
df1.head()
```

```python
# loc[] : 라벨 기반 인덱싱을 할 때 쓰는 메서드
## dates[0] : dates[1] : 행 선택(라벨, 조건식, 슬라이싱)
## 'E' : 열 선택(라벨 이름)
## = 1 : 선택된 구간(dates[0] : dates[1])의 열('E') 값을 1로 채움
df1.loc[dates[0] : dates[1], 'E']  = 1
df1.head()
```

```python
# isna() : NaN값이면 True, 아니면 False 출력
pd.isna(df1)
```

# 데이터 인덱싱과 필터링

## [] 컬럼 기반 필터링 또는 불린 인덱싱 필터링

### 1. []
- 컬럼 단일 선택 : df['col1'] -> Series 반환
- 컬럼 여러 개 선택 : df[['col1', 'col2']] -> DataFrame 반환
- Boolean Indexing과 함께 사용 가능 : df[df['col1'] > 10] -> 조건에 맞는 행 필터링

```python
# [] 연산자 내에 한 개의 컬럼만 입력하면 Series 객체를 반환
series = titanic['Name']
(series)
```

```python
# [] 연산자 내에 한 개의 컬럼을 리스트로 입력하면 DataFrame 반환
one_col_df = titanic[['Name']]
(one_col_df)
```

```python
# [] 연산자 내에 여러 개의 컬럼을 리스트로 입력하면 그 컬럼들로 구성된 DataFrame 반환
filtered_df = titanic[['Name', 'Age']]
filtered_df
```

### 2. loc[] / iloc[]
#### loc : 행 / 열 이름으로 접근(명칭(label) 기반 인덱싱)
- 컬럼(열)을 추출하는 경우 : '컬럼명' 같이 이름(label)으로 열 지정
- 행을 추출하는 경우 : 행의 인덱스 이름(label)을 사용해 지정
- 행과 열을 함께 지정하는 경우 -> df.loc[행이름, 열이름] 형식 사용

#### iloc : 정수 위치 기반 인덱싱 (숫자 인덱싱 사용) (위치(position) 기반 인덱싱)
- 레이블(이름)이 아닌 정수 위치로 데이터 지정
- 가로축(행), 세로축(열)의 좌표를 기반으로 행과 열 추출
- 행, 열의 위치값을 정수로 입력하여 선택 (0부터 시작)
  - 예 : `df.iloc[0, 1]` -> 첫 번째 행, 두 번쨰 열

### 3. Boolean Indexing
- 조건식에 따른 필터링

```python
import pandas as pd
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

```python
titanic.dtypes
```

```python
# 35살보다 나이가 많은 승객 추출
above_35 = titanic[titanic['Age'] > 35]
above_35.head()
```

```python
# class 2, 3 선실의 승객 추출하기
## isin(values) : Series나 DataFrame 전체에서 각 원소가 values 안에 들어있는지 확인
class_23 = titanic[titanic['Pclass'].isin([2, 3])]
class_23.head()
```

```python
class_23 = titanic[(titanic['Pclass'] == 2) | (titanic['Pclass'] == 3)]
class_23.head()
```

```python
# 나이가 알려진 승객 추출하기
## notna() : na값이 아닐 때 True, na값일 때 False
### 지금의 경우, na값이 아닐 때의 titanic 행 추출
age_no_na = titanic[titanic['Age'].notna()]
age_no_na.head()
age_no_na.shape
```

```python
# 35살 이상인 승객의 이름 추출하기
# loc[조건식, 열명]
adult_names = titanic.loc[titanic['Age'] > 35, 'Name']
adult_names.head()
```

```python
# 0번 ~ 3번 승객의 이름을 익명화하기
# iloc[행 범위, 열 번호] 선택
titanic.iloc[0:3, 3] = 'anonymous'
titanic.head()
```
